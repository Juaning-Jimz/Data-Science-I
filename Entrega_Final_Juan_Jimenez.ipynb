### **Bloque de limpieza de datos**

```python
# 1. Eliminar columnas poco √∫tiles o redundantes
df_clean = df.drop(columns=["jurisdiccion_residencia_nombre"], errors="ignore")

# 2. Imputar valores faltantes
df_clean["tipo_de_parto_nombre"] = df_clean["tipo_de_parto_nombre"].fillna("Desconocido")
df_clean["semana_gestacion"] = df_clean["semana_gestacion"].fillna("Sin especificar")

# 3. Crear variables num√©ricas para an√°lisis

# Edad media de la madre
def convertir_edad(grupo):
    if isinstance(grupo, str):
        grupo = grupo.strip()

        # Quitar prefijos como "3." o "5."
        if "." in grupo:
            grupo = grupo.split(".", 1)[1].strip()

        if "Menor" in grupo:
            return 14
        elif "45 y mas" in grupo or "45 y m√°s" in grupo:
            return 47
        elif "Sin especificar" in grupo:
            return None
        elif "a" in grupo:  # Ejemplo: "20 a 24"
            nums = [int(x.strip()) for x in grupo.split("a")]
            return sum(nums) / len(nums)
    return None

df_clean["edad_media"] = df_clean["edad_madre_grupo"].apply(convertir_edad)

# Semanas de gestaci√≥n (tomar el promedio de los intervalos)
def convertir_semanas(valor):
    if isinstance(valor, str):
        valor = valor.strip()
        if "Sin especificar" in valor:
            return None
        elif "Menos" in valor:
            return 21
        elif "m√°s" in valor or "y m√°s" in valor:
            return 43
        elif "a" in valor:  # Ejemplo: "32 a 36"
            nums = [float(x.strip()) for x in valor.split("a")]
            return sum(nums) / len(nums)
    return None

df_clean["semana_num"] = df_clean["semana_gestacion"].apply(convertir_semanas)

# 4. Mantener solo columnas relevantes
columnas_relevantes = [
    "anio", "edad_media", "semana_num", "nacimientos_cantidad",
    "tipo_de_parto_nombre", "Sexo", "instruccion_madre"
]
df_clean = df_clean[columnas_relevantes]

df_clean.head()
```

### **Outliers en boxplots comparativos antes y despu√©s de limpiarlos con IQR:**

```python
# Funci√≥n detecci√≥n de outliers con IQR
def detectar_outliers_iqr(serie):
    Q1 = serie.quantile(0.25)
    Q3 = serie.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (serie < lower_bound) | (serie > upper_bound)

# Detectar en edad_media y semana_num
outliers_edad = detectar_outliers_iqr(df_clean["edad_media"])
outliers_semana = detectar_outliers_iqr(df_clean["semana_num"])

print(f"Outliers en edad_media: {outliers_edad.sum()}")
print(f"Outliers en semana_num: {outliers_semana.sum()}")

# Crear dataset limpio
df_no_outliers = df_clean[~(outliers_edad | outliers_semana)]

print(f"Tama√±o original: {df_clean.shape[0]}")
print(f"Tama√±o sin outliers: {df_no_outliers.shape[0]}")
print(f"Registros eliminados: {df_clean.shape[0] - df_no_outliers.shape[0]}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 2, figsize=(12,8))

# 1. Edad de la madre con outliers
sns.boxplot(y=df_clean["edad_media"], ax=axes[0,0], color="skyblue")
axes[0,0].set_title("Edad media de la madre (con outliers)")

# 2. Edad de la madre sin outliers
sns.boxplot(y=df_no_outliers["edad_media"], ax=axes[0,1], color="lightgreen")
axes[0,1].set_title("Edad media de la madre (sin outliers)")

# 3. Semanas de gestaci√≥n con outliers
sns.boxplot(y=df_clean["semana_num"], ax=axes[1,0], color="skyblue")
axes[1,0].set_title("Semanas de gestaci√≥n (con outliers)")

# 4. Semanas de gestaci√≥n sin outliers
sns.boxplot(y=df_no_outliers["semana_num"], ax=axes[1,1], color="lightgreen")
axes[1,1].set_title("Semanas de gestaci√≥n (sin outliers)")

plt.tight_layout()
plt.show()
```

Conclusiones del an√°lisis de outliers

- **Edad de la madre:** los outliers corresponden a edades muy bajas o altas. Al quitarlos, la mayor√≠a de los partos se concentran entre **20 y 35 a√±os**.  
- **Semanas de gestaci√≥n:** aparecen valores inusuales (muy bajos o altos). Sin outliers, la distribuci√≥n queda centrada en el rango esperado.  
- **En general:** los outliers pueden distorsionar medidas estad√≠sticas, pero tambi√©n reflejan **casos cl√≠nicamente relevantes** como partos prematuros o embarazos en edades extremas.

```python
# Librer√≠as necesarias
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

## i) Selecci√≥n de caracter√≠sticas

Dado que el dataset contiene tanto variables num√©ricas como categ√≥ricas,
primero se aplica un proceso de **codificaci√≥n de etiquetas (Label Encoding)** para convertir las variables categ√≥ricas a formato num√©rico.

Posteriormente, se utiliza el m√©todo **SelectKBest** con la funci√≥n `f_regression`, que mide la correlaci√≥n estad√≠stica de cada variable independiente con la variable objetivo.

De este modo, se seleccionan las **6 variables m√°s relevantes** para la predicci√≥n.

```python
# === Cargar y preparar los datos ===
df = pd.read_csv("data-set (1).csv", encoding="latin1")
data = df.dropna()

y = data["nacimientos_cantidad"]
X = data.drop(columns=["nacimientos_cantidad"])

# Codificaci√≥n de variables categ√≥ricas
le_dict = {}
for col in X.select_dtypes(include=["object"]).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    le_dict[col] = le

# Divisi√≥n en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Selecci√≥n de las mejores caracter√≠sticas
selector = SelectKBest(score_func=f_regression, k=6)
X_train_sel = selector.fit_transform(X_train, y_train)
X_test_sel = selector.transform(X_test)

selected_features = X.columns[selector.get_support()]
print("Variables seleccionadas:", list(selected_features))
```

### Selecci√≥n de Caracter√≠sticas (Feature Selection)

Para reducir la dimensionalidad del dataset y eliminar variables redundantes o poco relevantes, se aplic√≥ el m√©todo **SelectKBest** con la funci√≥n estad√≠stica **f_regression**, que eval√∫a la relaci√≥n lineal entre cada variable independiente y la variable objetivo `nacimientos_cantidad`.

Este enfoque permite conservar √∫nicamente aquellas variables con mayor poder predictivo, optimizando as√≠ el rendimiento y la interpretabilidad del modelo.  
Adem√°s, se realiz√≥ un **an√°lisis de correlaci√≥n** para confirmar la coherencia entre las variables seleccionadas y la variable dependiente.


```python


from sklearn.feature_selection import SelectKBest, f_regression
import pandas as pd
import numpy as np

# Copiamos el dataset limpio (ajusta el nombre si tu DataFrame se llama distinto)
df = data.copy()

# Convertimos variables categ√≥ricas a num√©ricas
df_encoded = pd.get_dummies(df, drop_first=True)

# Separar variables independientes (X) y dependiente (y)
X = df_encoded.drop("nacimientos_cantidad", axis=1)
y = df_encoded["nacimientos_cantidad"]

# Aplicamos el m√©todo SelectKBest con funci√≥n de regresi√≥n
selector = SelectKBest(score_func=f_regression, k=10)  # selecciona las 10 m√°s relevantes
X_selected = selector.fit_transform(X, y)

# Obtenemos los nombres de las variables seleccionadas
selected_features = X.columns[selector.get_support()]
print("Variables seleccionadas por relevancia estad√≠stica:")
for col in selected_features:
    print(" -", col)

# Creamos un nuevo DataFrame con solo las columnas seleccionadas
X_reduced = X[selected_features]


import matplotlib.pyplot as plt
import seaborn as sns

corr = X_reduced.join(y).corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Mapa de correlaciones entre variables seleccionadas y cantidad de nacimientos")
plt.show()
```

## ii) Entrenamiento del modelo

Se utiliza un **Random Forest Regressor**, un modelo basado en m√∫ltiples √°rboles de decisi√≥n.

Las principales ventajas de este modelo son:
- Maneja tanto variables categ√≥ricas como num√©ricas.
- Tolera relaciones no lineales y ruido en los datos.
- Permite calcular la **importancia de las variables**.

El modelo se entrena con los datos seleccionados y se eval√∫a sobre un conjunto de prueba (30% del total).

```python
# Entrenamiento del modelo
model = RandomForestRegressor(random_state=42, n_estimators=100)
model.fit(X_train_sel, y_train)

# Predicciones
y_pred = model.predict(X_test_sel)

```

## iii) C√°lculo de m√©tricas de validaci√≥n

Para evaluar el desempe√±o del modelo se utilizan tres m√©tricas est√°ndar:

- **MAE (Error absoluto medio)**: promedio de las diferencias absolutas entre las predicciones y los valores reales.  
- **RMSE (Ra√≠z del error cuadr√°tico medio)**: penaliza m√°s los errores grandes.  
- **R¬≤ (Coeficiente de determinaci√≥n)**: indica la proporci√≥n de la varianza explicada por el modelo (1 es perfecto).

Estos indicadores permiten medir tanto la precisi√≥n promedio como la capacidad explicativa del modelo.


```python
# C√°lculo de m√©tricas
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("M√©tricas de desempe√±o del modelo:")
print(f"MAE  : {mae:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"R¬≤   : {r2:.3f}")
```

## iv) Importancia de las variables

El modelo de Random Forest permite obtener la **importancia relativa** de cada variable dentro del conjunto de caracter√≠sticas seleccionadas.  
Esto permite interpretar qu√© factores tienen mayor peso en la predicci√≥n de nacimientos.

```python
importances = model.feature_importances_

plt.figure(figsize=(8,5))
plt.barh(range(len(selected_features)), importances, align='center')
plt.yticks(range(len(selected_features)), selected_features)
plt.xlabel('Importancia')
plt.title('Importancia de las variables seleccionadas')
plt.show()
```
